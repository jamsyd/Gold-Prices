# Model for Gold

* There is a very well known relationship between Gold Prices and real interest rates
* This is the relationship I will investigate


```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.linear_model import LinearRegression 
import scipy.stats as stats
```

* The data I have used is from the Federal reserve of St. Louis data base

```python
data=pd.read_csv('gold.csv')

# This is our column of we use to predict

Y=data.iloc[1000:, 0].values.reshape(-1,1) # Gold

# This is the column we want to predict
X=data.iloc[1000:,1].values.reshape(-1,1) # TIP
```

# Plotting Gold Prices vs. TIPS yield

```python
## Plot residuals in the time domain
plt.figure(figsize=(25, 5))

plt.scatter(X,Y)
plt.xlabel("TIPS")
plt.ylabel("Gold Price")
plt.title("Gold vs. TIPS yield")

plt.show()
```

# Fitting the Model

```python
lr = LinearRegression()

# Fit the model
lr.fit(X,Y)

# Parameters generated by linear regression
a0 = lr.intercept_[0]
a1 = lr.coef_[0][0]

print(a0,a1)

gold_model=a0+a1*X
```

```python
plt.figure(figsize=(20, 5))

plt.plot(a0 + a1*X)
plt.plot(Y)
plt.legend(['Gold Price','Model'])
plt.ylabel("Prices")
plt.xlabel("Time")
plt.title("Gold Price vs. Model")

plt.show()

plt.figure(figsize=(20, 5))

plt.plot(stats.zscore(Y-(a0 + a1*X)))
plt.axhline(y=0,color='red')

plt.legend(['Spread between Gold Price and model'])
plt.ylabel("z-score")
plt.xlabel("Time")
plt.title("Spread between Gold Price and model")

plt.show()
```
